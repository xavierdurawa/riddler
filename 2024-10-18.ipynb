## Problem

_From Michael Lester comes a delectable dilemma about soup:_

_I have a large, hemispherical piece of bread with a radius of 1 foot. I make a bread bowl by boring out a cylindrical hole with radius r, centered at the top of the hemisphere and extending all the way to the flat bottom crust._

_What should the radius of my borehole be to maximize the volume of soup my bread bowl can hold?_

```python
# Visualizing

import matplotlib.pyplot as plt
import numpy as np

# Parameters for the semicircle
radius = 1
center = (0, 0)

# Create angles for the semicircle (0 to pi)
theta_angles = np.linspace(0, np.pi, 100)

# Parametric equations for the semicircle
x = center[0] + radius * np.cos(theta_angles)
y = center[1] + radius * np.sin(theta_angles)

# Create the plot
fig, ax = plt.subplots(figsize=(8,6))

# Plot the semicircle
ax.plot(x, y, label='Semicircle', color='blue')

# Define positions for the two vertical lines
x1 = -0.5  # First vertical line x-position
x2 = 0.5   # Second vertical line x-position

# Define y-limits for the vertical lines (from center to top of the semicircle)
y_start = center[1]
y_end = center[1] + radius

# Define a uniform color for both vertical lines
vertical_line_color = 'red'

# Plot the first vertical line
ax.plot([x1, x1], [y_start, y_end], label='Vertical Lines', color=vertical_line_color, linestyle='--')

# Plot the second vertical line
ax.plot([x2, x2], [y_start, y_end], color=vertical_line_color, linestyle='--')

# Select one vertical line to add 'r_c' and 'theta'
selected_x = x2  # You can choose x1 or x2
# Calculate the intersection point on the semicircle
intersection_y = np.sqrt(radius**2 - selected_x**2)

# Plot the line from intersection to origin
ax.plot([0, selected_x], [0, intersection_y], label='$r_c$', color='green', linewidth=2)

# Calculate angle theta in degrees for annotation
theta_deg = np.degrees(np.arctan2(intersection_y, selected_x))

# Add theta annotation
# Determine the position for the theta arc
theta_radius = 0.2
theta_theta = np.linspace(0, np.arctan2(intersection_y, selected_x), 100)
theta_x = theta_radius * np.cos(theta_theta)
theta_y = theta_radius * np.sin(theta_theta)
ax.plot(theta_x, theta_y, color='purple', linewidth=1.5)

# Add an arrow to indicate the angle
ax.annotate(
    '', 
    xy=(theta_radius * np.cos(theta_theta[-1]), theta_radius * np.sin(theta_theta[-1])),
    xytext=(0,0),
    arrowprops=dict(arrowstyle='->', color='purple')
)

# Place the theta label
mid_theta = len(theta_theta) // 2
ax.text(theta_x[mid_theta], theta_y[mid_theta], r'$\theta$', color='purple', fontsize=12, ha='left', va='bottom')

# Set equal scaling
ax.set_aspect('equal', 'box')

# Add grid, legend, and labels
ax.grid(True, linestyle='--', alpha=0.5)
ax.legend()
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Semicircle with Two Vertical Lines, $r_c$, and $\\theta$')

# Set limits for better visualization
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-0.1, 1.2)

# Show the plot
plt.show()
```

## Solution

We can model the volume of the soup as

$$ V = \pi r_c^2 h $$
$$ \text{ where } r_c = cos(\theta) \text{ and } h = sin(\theta) $$

where $r_c$ is the radius of the cylindrical hole and $h$ is the height of the cylindrical hole. We simply need to take the deriviative and set it to zero to find the maximum volume.

$$ \displaystyle \frac{dV}{d\theta} = \frac{d}{d\theta} [ \pi cos^2(\theta) sin(\theta) ] = \pi \left( cos^3(\theta) - 2sin^2(\theta)cos(\theta) \right) = \pi cos(\theta) \left( cos^2(\theta) - 2sin^2(\theta) \right) = 0 $$

Obviously the the $ cos(\theta) $ term is a degenrate solution, so we can ignore it. We are left with

$$ cos^2(\theta) - 2sin^2(\theta) = 0 $$

$$ tan(\theta) = \frac{1}{\sqrt{2}} $$

$$ cos(\theta) = \sqrt{\frac{2}{3}} \text{ and } sin(\theta) = \sqrt{\frac{1}{3}} $$

So $ r_c = \sqrt{\frac{2}{3}} $.
