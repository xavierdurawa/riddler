## Problem

_You ring the doorbell, and someone dressed as a mathematician answers. (What does a “mathematician” costume look like? Look in the mirror!) They present you with a giant bag from which to pick candy, and inform you that the bag contains exactly three peanut butter cups (your favorite!), while the rest are individual kernels of candy corn (not your favorite!)._

_You have absolutely no idea how much candy corn is in the bag—any whole number of kernels (including zero) seems equally possible in this monstrous bag._

_You reach in and pull out a candy at random (that is, each piece of candy is equally likely to be picked, whether it’s a peanut butter cup or a kernel of candy corn). You remove your hand from the bag to find that you’ve picked a peanut butter cup. Huzzah!_

_You reach in again and pull a second candy at random. It’s another peanut butter cup! You reach in one last time and pull a third candy at random. It’s the third peanut butter cup!_

_At this point, whatever is left in the bag is just candy corn. How many candy corn kernels do you expect to be in the bag?_

## Solution

Let's denote the number of candy corn kernels in the bag as $c$. The probability of drawing a peanut butter cup, _give_ that there are $c$ candy corn kernels in the bag, is 

$$ P(\text{draw 3 peanut butter cups} | C) = \frac{3}{3 + c} \cdot \frac{2}{2 + c} \cdot \frac{1}{1 + c} $$

Interestingly, in this problem, we are interested in the expected value of $c$ given that we have drawn three peanut butter cups. This is a classic example of a [posterior expectation](https://en.wikipedia.org/wiki/Posterior_expectation) problem. We can use Bayes' theorem to calculate the posterior distribution of $c$ given that we have drawn three peanut butter cups.

$$ P(c | \text{draw 3 peanut butter cups}) = \frac{P(\text{draw 3 peanut butter cups} | C) \cdot P(C)}{P(\text{3 peanut butter cups})} $$

We should note from the problm that the enormous bag makes any number of candy corn kernels equally likely. Therefore, the prior distribution of $c$ is uniform. For brevity, we will denote drawing three peanut butter cups as $D_3$.

$$ \displaystyle P(C | D_3) = \frac{P(D_3 | C) \cdot P(C)}{\displaystyle \sum_{c=0}^{\infty} P(D_3 | C=c) \cdot P(C)} $$

Since $ P(c) $ is uniform, we can simplify the equation to:

$$ \displaystyle P(C | D_3) = \frac{P(D_3 | C)}{\displaystyle \sum_{c=0}^{\infty} P(D_3 | C=c) } $$

since we want the expectation of $c$, we can calculate it as:

$$ E[C | D_3] = \frac{\displaystyle \sum_{c=0}^{\infty} c \cdot P(D_3 | C=c)}{\displaystyle \sum_{c=0}^{\infty} P(D_3 | C=c)} = \frac{\displaystyle \sum_{c=0}^{\infty} \frac{6c}{(3+c)(2+c)(1+c)}}{\displaystyle \sum_{c=0}^{\infty} \frac{6}{(3+c)(2+c)(1+c)}} = \frac{\displaystyle \sum_{c=0}^{\infty} \frac{c}{(3+c)(2+c)(1+c)}}{\displaystyle \sum_{c=0}^{\infty} \frac{1}{(3+c)(2+c)(1+c)}} $$

So how do we compute these infinite sums? We can use the [partial fraction decomposition](https://en.wikipedia.org/wiki/Partial_fraction_decomposition) to simplify the sums. Let's call the numerator $ N(c) $ and the denominator $ D(c) $.

$$ N(c) = \frac{c}{(3+c)(2+c)(1+c)} = \frac{A_1}{3+c} + \frac{A_2}{2+c} + \frac{A_3}{1+c} $$

$$ D(c) = \frac{1}{(3+c)(2+c)(1+c)} = \frac{B_1}{3+c} + \frac{B_2}{2+c} + \frac{B_3}{1+c} $$

Multiplying both sides by the denominator, we get:

$$ c = A_1(2+c)(1+c) + A_2(3+c)(1+c) + A_3(3+c)(2+c) $$

$$ 1 = B_1(2+c)(1+c) + B_2(3+c)(1+c) + B_3(3+c)(2+c) $$

Solving for $ A_1, A_2, A_3, B_1, B_2, B_3 $, we get:

$$ A_1 = -\frac{3}{2} , A_2 = 2, A_3 = -\frac{1}{2}, B_1 = \frac{1}{2}, B_2 = -1, B_3 = \frac{1}{2} $$

Now you might be think, _how does this help us?_ These terms are now of the form $ \sum \frac{1}{c} $, which we know is a harmonic and divergent. True, but if, in each case, the divergent part of each summation will cancel out, leaving us with a finite sum. This will hold true as long as as there is a constraint that $ A_1 + A_2 + A_3 = 0 $. Hopefully this becomes clearer as we calculate the sums. One other wrinkle we have to deal with is the fact that the sums start at different values, so we'll have to subtract some terms to make them all start at 1.
$$ N(c) = \displaystyle \sum_{c=0}^{\infty} -\frac{3}{2(3+c)} + \frac{2}{2+c} - \frac{1}{2(1+c)} = -\frac{3}{2} \sum_{c=0}^{\infty} \frac{1}{(3+c)} + 2 \sum_{c=0}^{\infty} \frac{1}{(2+c)} - \frac{1}{2} \sum_{c=0}^{\infty} \frac{1}{(1+c)} $$

Let's rewrite this a bit to avoid introducing the divergent terms:

$$ \lim_{n \to \infty} N(c) = -\frac{3}{2} H_{N+3} + 2 H_{N+2} - \frac{1}{2} H_{N+1} $$

$$ \lim_{n \to \infty} N(c) = -\frac{3}{2} (H_{N+1} - 1 - \frac{1}{2}) + 2 (H_{N+1} - 1) - \frac{1}{2} H_{N+1} = \frac{9}{4} - 2 = \frac{1}{4} $$

Now for the denominator term:

$$ D(c) = \displaystyle \sum_{c=0}^{\infty} \frac{1}{2(3+c)} - \frac{1}{(2+c)} + \frac{1}{2(1+c)} = \frac{1}{2} \sum_{c=0}^{\infty} \frac{1}{(3+c)} - \sum_{c=0}^{\infty} \frac{1}{(2+c)} + \frac{1}{2} \sum_{c=0}^{\infty} \frac{1}{(1+c)} $$

$$ \lim_{n \to \infty} D(c) = \frac{1}{2} H_{N+3} - H_{N+2} + \frac{1}{2} H_{N+1} $$

$$ \lim_{n \to \infty} D(c) = \frac{1}{2} (H_{N+1} - 1 - \frac{1}{2}) - (H_{N+1} - 1) + \frac{1}{2} H_{N+1} = \frac{1}{4} $$

So the expected value of $c$ is:

$$ E[C | D_3] = \frac{\frac{1}{4}}{\frac{1}{4}} = \boxed{1} $$

```python
import numpy as np
import matplotlib.pyplot as plt

def likelihood(c):
    if c < 0:
        return 0
    return (3 / (3 + c)) * (2 / (2 + c)) * (1 / (1 + c))

def proposal(c):
    return c + np.random.randint(-1, 2)  # Propose a new state by adding -1, 0, or 1

def metropolis_hastings(n_samples, initial_state):
    samples = np.zeros(n_samples)
    samples[0] = initial_state
    for i in range(1, n_samples):
        current_state = samples[i-1]
        proposed_state = proposal(current_state)
        
        acceptance_ratio = likelihood(proposed_state) / likelihood(current_state)
        acceptance_probability = min(1, acceptance_ratio)
        
        if np.random.rand() < acceptance_probability:
            samples[i] = proposed_state
        else:
            samples[i] = current_state
    
    return samples

# Parameters
n_samples = 1000000
initial_state = 1  # Start with an initial guess

# Run Metropolis-Hastings
samples = metropolis_hastings(n_samples, initial_state)

# Discard burn-in period
burn_in = 50000
samples = samples[burn_in:]

# Calculate the expected number of candy corn kernels
expected_candy_corn = np.mean(samples)
print(f"Expected number of candy corn kernels: {expected_candy_corn}")

# Plot the results
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g')
plt.title('Posterior Distribution of Candy Corn Kernels')
plt.xlabel('Number of Candy Corn Kernels')
plt.ylabel('Density')
plt.show()
```

## Extra Credit

_You move on to the next house in your neighborhood, which is even more mysterious._

_The owner of this house presents you with a similarly large bag, which they say has exactly N peanut butter cups, where N ≥ 3. The remainder of the bag is filled with some unknown quantity of candy corn kernels, with any amount of them again being equally likely._

_You reach into the bag k times, with 3 ≤ k ≤ N, and pull out a candy at random. Each time, it’s a peanut butter cup!_

_How many candy kernels do you expect to be in the bag?_

## Solution

So most of the solution above still holds, but we need to adjust the sums to account for the fact that we are drawing $k$ peanut butter cups out of a possible $N$. The sums will now start at $N - k + 1$.

$$ E[C | D_k] = \frac{\displaystyle \sum_{c=0}^{\infty} c \cdot P(D_k | C=c)}{\displaystyle \sum_{c=0}^{\infty} P(D_k | C=c)} = \frac{\displaystyle \sum_{c=0}^{\infty} \frac{\frac{N!}{(N-k)!} c}{(N+c)(N-1+c)...(N-k+1+c)}}{\displaystyle \sum_{c=0}^{\infty} \frac{\frac{N!}{(N-k)!}}{(N+c)(N-1+c)...(N-k+1+c)}} = \frac{\displaystyle \sum_{c=0}^{\infty} \frac{c}{(N+c)(N-1+c)...(N-k+1+c)}}{\displaystyle \sum_{c=0}^{\infty} \frac{1}{(N+c)(N-1+c)...(N-k+1+c)}} $$

Using product notation:

$$ E[C | D_k] = \frac{\displaystyle \sum_{c=0}^{\infty} \frac{c}{\displaystyle \prod_{i=0}^{k-1} (N-i+c)}}{\displaystyle \sum_{c=0}^{\infty} \frac{1}{\displaystyle \prod_{i=0}^{k-1} (N-i+c)}} $$

We can use the same partial fraction decomposition to simplify the sums. Let's call the numerator $ N(c) $ and the denominator $ D(c) $. A single term of the numerator is:

$$ \frac{c}{\displaystyle \prod_{i=0}^{k-1} (N-i+c)} \rightarrow c = \displaystyle \sum_{i=0}^{k-1} \frac{A_i}{N-i+c} $$

To solve for $ A_i $ let's use the method of residues.

$$ \lim_{c \to -N+i} (c+N-i) \frac{c}{\displaystyle \prod_{j=0}^{k-1} (N-j+c)} = A_i $$

$$ A_i = \frac{c}{\displaystyle \prod_{j=0 \\ j \ne i}^{k-1} (N-j+c)} \bigg|_{c= -(N-i)} = \frac{-(N-i)}{\displaystyle \prod_{j=0 \\ j \ne i}^{k-1} (j-i)} $$

How do we compute this denominator? We will first need to consider when it is negative vs positive and from there, we can express it as factorials. Each term where $ i \gt j $, there will be a negative term that needs to be accounted for, for which there will be $ i $ of them from the denominator and $ k $ of them from the numerator.  Otherwise, the positive terms in the denominator will just be $ i! $ and the negative terms will have magnitude $ (k-1-i)! $

$$ A_i = \frac{N-i}{(-1)^{k+i} \cdot i! \cdot (k-1-i)!} $$

For the denominator, its nearly the same.

$$ \lim_{c \to -N+i} (c+N-i) \frac{1}{\displaystyle \prod_{j=0}^{k-1} (N-j+c)} = A_{D_i} $$

$$ A_{D_i} = \frac{1}{\displaystyle \prod_{j=0 \\ j \ne i}^{k-1} (N-j+c)} \bigg|_{c= -(N-i)} = \frac{1}{\displaystyle \prod_{j=0 \\ j \ne i}^{k-1} (i-j)} $$

$$ A_{D_i} = \frac{1}{(-1)^{k-i-1} \cdot i! \cdot (k-1-i)!} $$

Returning to the computation of the sum

$$ N(N,k) = \displaystyle \sum_{c=0}^{\infty} \left( \sum_{i=0}^{k-1} \frac{A_{N_i}}{N-i+c} \right) = \displaystyle \sum_{i=0}^{k-1} \left( \sum_{c=0}^{\infty} \frac{A_{N_i}}{N-i+c} \right) = \displaystyle \sum_{i=0}^{k-1} \left( A_{N_i} \sum_{c=0}^{\infty} \frac{1}{N-i+c} \right) $$

Rewriting in terms of Harmonic terms

$$ N(N,k) = \displaystyle \sum_{i=0}^{k-1} A_{N_i} (H_{\infty} - H_{N-i}) $$

We just need to ensure that the sum of all $ A_{N_i} $ (and $ A_{D_i} $) is $ 0 $. This will always be true as long as the degree of $ \displaystyle \prod_{i=0}^{k-1} (N-i+c) $ is greater than degree 1. 

And we're left with

$$ E[C | D_k] = \frac{N(N,k)}{D(N,k)} = \frac{\displaystyle \sum_{i=0}^{k-1} A_{N_i} H_{N-i}}{\displaystyle \sum_{i=0}^{k-1} A_{D_i} H_{N-i}} $$

where $ H_i $ is the $ i \text{-th} $ harmonic number and $ A_{N_i} = \frac{N-i}{(-1)^{k+i} \cdot i! \cdot (k-1-i)!} $ and $ A_{D_i} = \frac{1}{(-1)^{k-i-1} \cdot i! \cdot (k-1-i)!} $

Below is a function that calculates the expected value of $ c $ given $ N $ and $ k $.

```python
import numpy as np

def compute_expectation(N, k):
    def harmonic(n):
        if n <= 0:
            return 0
        return sum(1 / i for i in range(1, n+1))
    
    def A_N_i(i):
        return (N - i) / ((-1)**(k+i) * np.math.factorial(i) * np.math.factorial(k-1-i))
    
    def A_D_i(i):
        return 1 / ((-1)**(k+i-1) * np.math.factorial(i) * np.math.factorial(k-1-i))
    
    numerator = sum(A_N_i(i) * harmonic(N-i-1) for i in range(k))
    denominator = sum(A_D_i(i) * harmonic(N-i-1) for i in range(k))
    
    return numerator / denominator

print(f"Expected number of candy corn kernels: {compute_expectation(4, 4)}")
```

```python
import sympy as sp

def compute_expectation_symbolic(N, k):
    def harmonic(n):
        if n <= 0:
            return 0
        return sum(sp.Rational(1, i) for i in range(1, n+1))
    
    def A_N_i(i):
        return (N - i) / ((-1)**(k+i) * sp.factorial(i) * sp.factorial(k-1-i))
    
    def A_D_i(i):
        return 1 / ((-1)**(k+i-1) * sp.factorial(i) * sp.factorial(k-1-i))
    
    A_N_i_values = [A_N_i(i) for i in range(k)]
    A_D_i_values = [A_D_i(i) for i in range(k)]
    harmonic_values = [harmonic(N-i) for i in range(k)]
    
    numerator = sum(A_N_i(i) * harmonic(N-i-1) for i in range(k))
    denominator = sum(A_D_i(i) * harmonic(N-i-1) for i in range(k))
    
    return numerator / denominator

N, k = 10, 3
expected_candy_corn = compute_expectation_symbolic(N, k)
print(f"Expected number of candy corn kernels: {expected_candy_corn}")
```

```python
from matplotlib import pyplot as plt

# Plot the values for various values of N and k (N >= k)
N_values = range(3, 20)
k_values = range(3, 20)
# ensure N >= k
expected_values = np.array([[compute_expectation(N, k) if N >= k else 0 for k in k_values] for N in N_values])

plt.figure(figsize=(10, 6))
masked_values = np.ma.masked_where(expected_values == 0, expected_values)
cmap = plt.cm.viridis
cmap.set_bad(color='gray')

plt.imshow(masked_values, cmap=cmap, interpolation='nearest', origin='lower')
plt.colorbar(label='Expected Value')
plt.title('Expected Number of Candy Corn Kernels')
plt.xlabel('k')
plt.ylabel('N')
plt.xticks(range(len(k_values)), k_values)
plt.yticks(range(len(N_values)), N_values)
plt.show()
```
