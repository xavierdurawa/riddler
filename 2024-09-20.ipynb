# Problem

_Suppose you have the parallelogram of cardboard shown below, which has side lengths of 2 units and 6 units, and angles of 30 degrees and 150 degrees:_


_By swirling two edges together, it’s possible to neatly (without any overlap) generate the lateral surface of a right cylinder—in other words, a toilet paper roll! (If you’re not convinced, try gently tearing a toilet paper roll along its diagonal seam and then unwrapping it into a flat shape. You get a parallelogram!)_

_Determine the volume of a cylinder you can make from this particular piece of cardboard._

# Solution

There are two possible orientations for the parallelogram, depending on which pair of sides are glued together. In the first orientation, the sides of length 2 are glued together, and in the second orientation, the sides of length 6 are glued together. We will calculate the volume of the cylinder for each orientation and take the larger of the two volumes.

In the first orientation, the height of the cylinder is $l_1$ and the circumference is 2. $l_1$ is simply $6sin(30) = 3$ and the radius is $\frac{1}{\pi}$. The volume of the cylinder is $\pi r^2 h = \pi \frac{1}{\pi^2} 3 = \frac{3}{\pi}$.

For the second orientation, the height of the cylinder is $l_2$ and the circumference is 6. $l_2$ is simply $2sin(30) = 1$ and the radius is $\frac{3}{\pi}$. The volume of the cylinder is $\pi r^2 h = \pi \frac{9}{\pi^2} 1 = \frac{9}{\pi}$.

So the second orientation is larger and our answer is $\boxed{\frac{9}{\pi}}$.

# Extra Credit

_Suppose you have a parallelogram with an area of 1 square unit. Let V represent the average volume of all cylinders whose lateral surface you can neatly make by swirling two edges of the parallelogram together._

_What is the minimum possible value of V?_

# Solution to Extra Credit

Let's first find the general expression for the volume of a cylinder given a parallelogram with area 1. Let $s_1$ and $s_2$ be the sides of the parallelogram, and let $\theta$ be the angle between them. to make things easier, let's also calculate $l_1$ and $l_2$ in terms of $s_1$, $s_2$, and $\theta$. We have:

$$l_1 = s_1sin(\theta)$$
$$l_2 = s_2sin(\theta)$$

We of course also have the constraint that the area of the parallelogram is 1:

$$ s_1l_2 = s_1s_2sin(\theta) = 1$$

If we glue the sides of length $s_1$ together, the height of the cylinder is $l_1$ and the circumference is $s_2$ making the radius $\frac{s_2}{2\pi}$. The volume of the cylinder is:

$$\pi r^2 h = \pi \left(\frac{s_2}{2\pi}\right)^2 s_1sin(\theta) = \frac{s_2^2s_1sin(\theta)}{4\pi}$$

If we glue the sides of length $s_2$ together, the height of the cylinder is $l_2$ and the circumference is $s_1$ making the radius $\frac{s_1}{2\pi}$. The volume of the cylinder is:

$$\pi r^2 h = \pi \left(\frac{s_1}{2\pi}\right)^2 s_2sin(\theta) = \frac{s_1^2s_2sin(\theta)}{4\pi}$$

To get the average volume, we need to average these two volumes.

$$V = \frac{1}{2}\left(\frac{s_2^2s_1sin(\theta)}{4\pi} + \frac{s_1^2s_2sin(\theta)}{4\pi}\right) = \frac{s_1s_2sin(\theta)}{8\pi}(s_1 + s_2)$$

Imposing the constraint that the area of the parallelogram is 1, we have:

$$s_1 = \frac{1}{s_2sin(\theta)}$$

Plugging that into the expression for $V$, we have:

$$V = \frac{1}{8\pi} \left( \frac{1}{s_2sin(\theta)} + s_2 \right) $$

We want to maximize $sin$ since it appears in the denominator and we know $sin$ has maximum value of $1$ at $\theta = 90$

$$V = \frac{1}{8\pi} \left( \frac{1}{s_2} + s_2 \right) $$

So if we just take a deriviative with respect $s_2$ and set it to 0, we get:

$$\frac{dV}{ds_2} = \frac{1}{8\pi} \left( -\frac{1}{s_2^2} + 1 \right) = 0$$

$$s_2^2 = 1$$

$$s_2 = 1$$

So the volume is minimized when our 'parallelogram' is actually a normal square!
